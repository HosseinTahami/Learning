<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        
| پارامتر| توضیحات|
| --------|------- |
| file_path| آدرس فایل یا URL فایلی که می‌خواهیم از آن داده را بخوانیم. آدرس‌دهی می‌تواند نسبی یا مطلق باشد.|
| sep| هر ستون را با این کاراکتر از ستون‌های دیگر جدا می‌کند. چون قرار است CSV بخوانیم، به طور پیش‌فرض مقدار `sep` کاراکتر ‍`,` است. اما ممکن است حتی فایل‌هایی که پسنود csv دارند با کاما از یکدیگر جدا نشده باشند!|
| header | این پارامتر شماره سطری را نشان می‌دهد که اسم ستون‌ها در آن آمده است. به طور پیش‌فرض روی حالت `infer` قرار دارد؛ یعنی اولین سطر را به عنوان اسم ستون لحاظ می‌کند. حالت `infer` معادل `header=0` است.<br>می‌تواند عدد صحیح، لیستی از اعداد صحیح یا `None` باشد.<br><br>- اگر یک عدد صحیح باشد، شماره سطری که اسامی ستون‌ها در آن‌جا قرار دارد را نشان می‌دهد.<br>- اگر لیستی از اعداد صحیح باشد، منجر به ایجاد دیتافریم‌هایی با چندین نمایه می‌شود که این مبحث خارج از مباحث این دوره است.<br>- اگر `None` باشد، هیچ اسمی برای ستون‌ها انتخاب نمی‌شود و پانداز به صورت پیش‌فرض ستون‌ها را از 0 تا C - 1 اسم‌گذاری می‌کند که C تعداد ستون‌هاست. |
| squeeze| اگر فایلی که می‌خوانیم یک ستون داشته باشد، می‌توانیم در قالب یک دیتاسری خروجی بگیریم. نوع مقادیر این پارامتر `bool‍` است. <br><br>در صورتی که فایل مورد نظر تنها دارای یک ستون باشد :<br>- اگر مقدار این پارامتر False باشد، خروجی `read_csv` دیتافریم است.<br>- اگر True باشد، خروجی `read_csv` دیتاسری است.|
|prefix| فرض کنید به طور پیش‌فرض اسم ستون‌های یک دیتافریم اعداد 0 تا C -1 هستند که C تعداد ستون‌هاست. اگر بخواهیم قبل از اعداد یک رشته اضافه کنیم به طوری‌که فرضا اسم ستون‌ها X0, X1, X2, … شود می‌توانیم رشته مورد نظر را در پارامتر prefix قرار دهیم. در حالتی که اسم ستون‌ها عدد نیستند هم می‌توانیم prefix داشته باشیم.|
| error_bad_lines | برخی سطر‌ها ممکن است معیوب باشند. به طور مثال اگر دیتافریم ۱۰ ستون دارد، سطری ۱۱ ستون داشته باشد (یک کاما اضافه‌تر داشته‌باشد). در حالت پیش‌فرض، پانداز برای خواندن این فایل‌ها دچار error می‌شود. اگر مقدار این پارامتر را برابر False قرار دهیم، پانداز سطرهای معیوب را از دیتافریمی که خروجی می‌دهد، حذف می‌کند. اگر مقدار این پارامتر را تنظیم نکنیم، عملیات با خطا روبرو شده و هیچ دیتافریمی خروجی نمی‌گیریم.|
|nrows| در شرایطی ممکن است نیاز باشد تنها تعدادی از سطر‌های یک فایل را بخوانیم؛ مثلا مواقعی که فایل حجیم باشد! <br>برای تعیین تعداد سطر‌هایی که می‌خواهیم از یک فایل بخوانیم از این پارامتر استفاده می‌کنیم. اگر مقدار 1000 داشته باشد، تنها 1000 سطر اول از فایل را دریافت می‌کنیم.|
</font>

    <code>read_html</code> پارامترهای مهم تابع

```python
pd.read_html(source, match='.+', flavor=None, header=None)
```
            



<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        
| پارامتر | توضیحات|
| ------- | --------|
| source  | منبعی از جنس HTML که قصد داریم جداول آن‌ را استخراج کنیم. منبع می‌تواند در قالب‌های زیر باشد.<br><br>- آدرس یک فایل HTML در حافظه محلی. مثلا `home/pandas/index.html‍` <br>- آدرس یک فایل HTML در فضای وب.<br>- یک رشته که درون آن کد HTML نوشته شده باشد. در مثال‌ این نمونه را بیشتر بررسی خواهیم کرد. |
| match   | ممکن است در یک فایل HTML جدول‌های زیادی وجود داشته باشد. برای اینکه بتوانیم راحت‌تر جدول مورد نظر خود را پیدا کنیم می‌توانیم یک رشته یا یک عبارت منظم (Regex) متناظر با جداول مد نظر خود را در این پارامتر قرار دهیم. در مثال‌ این قسمت بیشتر با کاربرد پارامتر `match` آشنا می‌شویم.                                                                                        |
| flavor  | برای تجزیه کردن (parse) HTML ورودی، موتورهای مختلفی وجود دارد. به طور پیش‌فرض pandas از lxml برای تجزیه HTML و تبدیلش به دیتافریم استفاده می‌کند. اگر استفاده از lxml دچار مشکل شود سراغ موتورهای `bs4` و `html5lib` می‌رود. اگر نتواند هیچ کدام از این موتورها را استفاده کند، اِرُر می‌دهد و در متن اِرُر راهنمایی می‌کند که کدام موتور را می‌توانیم نصب کنیم!                    |
| header  | این پارامتر، شماره سطری را نشان می‌دهد که اسم ستون‌ها در آن آمده است. توضیحات تکمیلی را می‌توانید از جدول‌های قبلی مطالعه کنید.                                                                                                                                                                                                                                                 |


</div>

````python
caption = "Racial composition in 2020"

df = pd.read_html('https://en.wikipedia.org/wiki/Minnesota', match=caption)[0]
df
````

<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        <b>مثال تابع read_html</b>
        <br>
کد یک جدول به HTML در رشته source نوشته شده است. جدول را در دیتافریمی بارگذاری کنید و سطرهایی را از دیتافریم انتخاب کنید که ستون Age در آن‌ها کمترمساوی 24 باشد.
</div>
    

````python
source = """
    <!DOCTYPE html>
<html>
<head>
</head>
<body>
<table>
  <tr>
    <th>Name</th>
    <th>Family</th>
    <th>Age</th>
  </tr>
  <tr>
    <td>Sajjad</td>
    <td>Yazdanparast</td>
    <td>23</td>
  </tr>
  <tr>
    <td>Parto</td>
    <td>Shahroodi</td>
    <td>25</td>
  </tr>
  <tr>
    <td>Hadi</td>
    <td>Sotude</td>
    <td>28</td>
  </tr>
  <tr>
    <td>Hamed</td>
    <td>Maansoury</td>
    <td>25</td>
  </tr>
  <tr>
    <td>SeyedParsa</td>
    <td>Mirtaheri</td>
    <td>24</td>
  </tr>
  <tr>
    <td>MohammadMahdi</td>
    <td>Shokri</td>
    <td>24</td>
  </tr>
  <tr>
    <td>MohammadBagher</td>
    <td>Tabrizi</td>
    <td>28</td>
  </tr>
</table>
</body>
</html>
"""

df = pd.read_html(source)[0]
df[df.Age <= 24]
````
