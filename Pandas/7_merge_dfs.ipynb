```python
import pandas as pd
import numpy as np
cars_dictionary = {
    "model" :["Peykan", "Peykan", "Peykan", "BMW X6", "BMW X6", "Nissan Junior", "Nissan Junior", "Nissan Junior", "Nissan Junior"],
    "Color" :["Red", "White", "Olive",  "Black", "Ametrine", "Blue", "White", "Blue", "Red"],
    "HP" :[48, 48, 48, 335, 335, 93, 93, 93, 93],
    "Weight (KG)" :[990, 990, 990, 2320, 2320, 1540, 1540, 1540, 1540],
    "year" :[1346, 1357, 1388, 1391, 1393, 1375, 1383, 1375, 1389]
}

cars_df = pd.DataFrame(cars_dictionary,
                      index = [14234, 44234, 22312, 42348, 16373, 38561, 58193, 85612, 62471])
cars_df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
     می‌خواهیم برای دیتافریم بالا، قیمت تعیین کنیم.
    </font>
</p>


```python
price_dictionary = {
    "model" : ["Peykan", "BMW X6", "Nissan Junior"],
    "price" : [40, 5200, 150]
}
price_df = pd.DataFrame(price_dictionary)
price_df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
دیتافریم دوم، با توجه به مدل هر ماشین قیمت‌ها را مشخص می‌کند.
    </font>
</p>


```python
pd.merge(cars_df, price_df)
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
می‌بینید که ستون جدیدی با نام <code>price</code>، به دیتافریم اضافه شده و اطلاعات مربوط هر ماشین را با تطبیق دادن ستون <code>model</code> قیمت گذاری کرده است. دقت کنید که در این حالت، خود تابع <code>pd.merge</code>، با پیدا کردن ستونی که نام مشترکی در هر دو دیتافریم داشت (<code>model</code>) این ستون را به عنوان کلید برای ادغام انتخاب کرده است.
    </font>
</p>

```python
new_df = cars_df.reset_index()
new_df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
از تابع <code>reset_index</code>، برای اعمال نمایه‌های جدید و ذخیره‌ی نمایه‌ی قبلی به عنوان ستون جدید استفاده می‌کنیم.
    </font>
</p>

```python
pd.merge(new_df, price_df)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این حالت، اطلاعات مربوط به شماره‌ی فاکتور که همان نمایه‌های قبلی بودند را از دست نمی‌دهیم.
    </font>
</p>

```python
price_dictionary2 = {
    "car model" : ["Peykan", "BMW X6", "Nissan Junior"],
    "price" : [40, 5200, 150]
}
price_df2 = pd.DataFrame(price_dictionary2)
price_df2
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
این بار، دیتافریمی که حاوی قیمت‌هاست، مدل ماشین‌ها را در ستونی با نام <code>car model</code> ذخیره می‌کند بنابراین، تابع <code>merge</code>، نمی‌تواند خودکار کلید را تشخیص بدهد.
    </font>
</p>

```python
merged = pd.merge(cars_df, price_df2, left_on = "model", right_on = "car model") 
merged
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
از پارامترهای <code>left_on</code> و <code>right_on</code>، برای مشخص کردن کلیدها استفاده می‌کنیم
    </font>
</p>

```python
merged.drop(columns = ["car model"])
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
تابع <code>drop</code> را برای حذف ستون با نام <code>car model</code> استفاده کرده‌ایم.
<br>
نحوه‌ی دقیقتر کار با این تابع را در فصل بعدی یاد می‌گیرید
    </font>
</p>

```python
price_with_year_dictionary = {
    "year" :[1346, 1357, 1388, 1391, 1393,  1375, 1383, 1389],
    "car model" :["Peykan", "Peykan", "Peykan", "BMW X6", "BMW X6", "Nissan Junior", "Nissan Junior", "Nissan Junior"],
    "price" :[25, 30, 48, 2200, 4800, 105, 130, 155]
}

price_with_year_df = pd.DataFrame(price_with_year_dictionary)
price_with_year_df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این حالت، می‌خواهیم از دو ستون سال ساخت و مدل ماشین به عنوان کلید ادغام استفاده کنیم
    </font>
</p>


```python
pd.merge(cars_df, price_with_year_df, left_on = ["model" , "year"], right_on = ["car model", "year"])
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
پارامترهای <code>left_on</code> و <code>right_on</code>، برای استفاده از چند ستون به عنوان کلید نیز، کار ما را راحت می‌کنند.
    </font>
</p>

```python
price_dictionary3 = {
    "HP" :[40, 330, 87],
    "price" :[48, 2200, 105]
}

price_df3 = pd.DataFrame(price_dictionary3,
                        index = [ "Peykan", "BMW X6", "Nissan Junior"])
price_df3
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
دیتافریم <code>price_df3</code>، نام مدل ماشین‌ها را به عنوان نمایه‌ی خود دارد.
    </font>
</p>

```python
pd.merge(cars_df, price_df3, left_on = "model" , right_index = True)
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای استفاده از نمایه‌های دیتافریم <code>price_df3</code>، کافی است تا از پارامتر <code>right_index = True</code> استفاده کنیم.
<br>
همان‌طور که در خروجی بالا می‌بینید، ستون <code>HP</code>، در هر دو دیتافریم وجود داشت بنابراین خود پانداس دو ستون را در خروجی در نظر گرفته اما برای مشخص شدن تفاوت آن‌ها از پسوندهای x  و y استفاده می کند.
    </font>
</p>

```python
pd.merge(cars_df, price_df3, left_on = "model" , right_index = True, suffixes=('_gallery', '_web'))
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای تغییر پسوندهای ستون‌های تکراری، می‌توانیم از پارامتر <code>suffixes</code> استفاده کنیم.
<br>
در این حالت، برای دیتافریم اول (<code>cars_df</code>)، پسوند <code>_gallery</code> در نظر گرفته می‌شود و برای دیتافریم دوم، (<code>price_df3</code>)، پسوند <code>_web</code> نوشته می‌شود.
    </font>
</p>

```python
full_cars_dictionary = {
    "model" :["Peykan", "Peykan", "Peykan", "BMW X6", "Nissan Junior", "Pride 131", "Pride 131"],
    "Color" :["Red", "White", "Olive", "White", "Black", "Ametrine", "Blue"],
    "HP" :[48, 48, 48, 335, 93, 71, 71],
    "Weight (KG)" :[990, 990, 990, 2320, 1540, 915, 915]
}

full_cars_df = pd.DataFrame(full_cars_dictionary)
full_cars_df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در دیتافریم بالا، تمامی ماشین‌های نمایشگاه را ثبت کرده‌ایم.
    </font>
</p>

```python
full_price_dictionary = {
    "model" :["Peykan", "BMW X6", "Nissan Junior", "Dena"],
    "price" :[37, 4800, 105, 380]
}

full_price_df = pd.DataFrame(full_price_dictionary)
full_price_df
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
دیتافریم <code>full_price_df</code>، قیمت تمامی ماشین‌های موجود در بازار را نشان می‌دهد. برخی از این ماشین‌ها ممکن است در نمایشگاه موجود نباشند.
    </font>
</p>

```python
pd.merge(full_cars_df, full_price_df)
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
با اعمال عملیات ادغام در حالت عادی، سطرهایی که مدل مشترک نداشتند حذف می‌شوند. برای مثال ماشین‌های پراید در دیتافریم اولیه و ماشین دنا در دیتافریم قیمت‌ها در خروجی ظاهر نشده است.
<br>
دلیل این نتیجه، حالت پیش‌فرض تابع <code>merge</code> است. در این جا نوع عملیات ادغام، به صورت <code>inner_join</code> تعریف شده است که به معنی حذف موردهای غیر مشترک است.
    </font>
</p>

```python
pd.merge(full_cars_df, full_price_df, how = "left")
```


 <p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای آن‌که تمامی اطلاعات دیتافریم اولیه (<code>full_cars_df</code>) بدون در نظر گرفتن اشتراک داشتن یا نداشتن با دیتافریم قیمت‌ها در نتیجه ظاهر شود، از پارامتر <code>how = "left"</code> استفاده می‌کنیم. به این صورت حالت عملیات ادغام به صورت <code>left-join</code> تعریف می‌شود.
<br>
 پیشنهاد می‌کنیم برای یادگیری بیشتر، حالات <code>right</code> و <code>outer</code> را نیز امتحان کنید.
    </font>
</p>
