<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در این درس، سعی می‌کنیم انواع روش‌های بررسی دیتافریم‌هایی که دارای داده‌های زیادی هستند را با هم مرور کنیم.
    </font>
</p>

```python
import pandas as pd
import numpy as np

df = pd.DataFrame([['Peykan', 'Red', 48, 990, 33],
       ['BMW X6', 'White', 335, 2320, 2400],
       ['Peugeot 206', 'Black', 105, 1025, 105],
       ['Tiba', 'White', 87, 1027, 120],
       ['Nissan Junior', 'Blue', 93, 1540, 230],
       ['Nissan Junior', 'Red', 93, 1540, 186],
       ['renault 5', 'Silver', 48, 850, 38]],
       index = ['a','b','c','d','e','f','g'],
       columns= ['model' , 'Color' , 'HP', 'Weight (KG)', 'price'])
df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در کد بالا، یک دیتافریم را با استفاده از آرایه نامپای تعریف کرده‌ایم.
</font>
</p>

```python
df.head()
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    با استفاده از تابع <code>()head</code>، به ۵ سطر اول دیتافریم دسترسی پیدا می‌کنیم.
</font>
</p>

```python
df.tail()
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    با استفاده از تابع <code>()tail</code>، به ۵ سطر آخر دیتافریم دسترسی پیدا می‌کنیم.
</font>
</p>

```python
df.head(6)
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    برای دسترسی به بیشتر از ۵ سطر در توابع <code>()head</code> و <code>()tail</code>، می‌ةوان تعداد سطر‌هایی که مورد انتظارمان است را به عنوان پارامتر به تابع بدهیم.
<br>
در کد بالا این مثال برای تابع <code>()head</code> زده شده است اما پیشنهاد می‌کنیم برای آشنایی بیشتر آن را روی تابع <code>()tail</code> نیز امتحان کنید.
</font>
</p>

```python
df.sample(4 , random_state = 42)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    با استفاده از کد بالا، ۴ سطر تصادفی از دیتافریم را انتخاب کرده‌ایم.
</font>
</p>

```python
df.sample(frac=0.4 , random_state = 42)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در صورتی که بخواهیم به جای مشخص کردن تعداد سطر‌های نمونه، درصدی از آن‌ها را انتخاب کنیم می‌توانیم از پارامتر <code>frac</code> برای این کار استفاده کنیم.
</font>
</p>

```python
df.sample(frac=0.4, ignore_index=True , random_state = 42)
```


 <p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    پارامتر <code>ignore_index</code>، برای ریست کردن نمایه‌ها بعد از انتخاب تصادفی استفاده می‌شود.
<br>
    اگر بخواهیم دید کلی‌تر و دقیق‌تری بر دیتافریم داشته باشیم می‌توانیم از تابع <code>()info</code> و <code>()describe</code> نیز استفاده کنیم.
</font>
</p>

```python
df.info()
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    خروجی تابع <code>()info</code>، تمامی ساختار دیتافیم را برای ما نشان می‌دهد. می‌بینیم که نام نمایه‌ها را در خط دوم توضیح داده است که به صورت ۷ نمایه از a تا g می‌باشد. همچنین برای تمامی ستون‌ها، می‌توانیم نوع مقادیر Dtype را مشاهده کنیم. به طور معمول مقادیر عددی را به صورت int یا float در نظر می‌گیریم مگر آن‌که این ستون حاوی مقادیر خالی (null) باشد. نحوه‌ی کار با این مقایر خالی را در درسنامه‌ی «کار با مقادیر خالی» در فصل آماده‌سازی یاد می‌گیرید. می‌بینیم که این دیتافریم دقیقا حاوی ۷ داده است که تمامی ستون‌ها ۷ داده‌ی غیرخالی دارند پس می‌توان از مقادیر آن مطمئن بود و بدون مشکل به باقی محاسبات ادامه داد.
</font>
</p>

```python
df.describe()
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    همان‌طور که می‌بینید، این تابع تنها برای ستون‌هایی که مقادیر عددی داشتند اعمال شده و دید دقیقی از باقی ستون‌ها به دست نمی‌آوریم.
</font>
</p>

```python
df['Color'].value_counts()
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    یکی از راه‌هایی که برای آشنایی با ستون‌های دارای داده‌ی دسته‌ای (categorical) وجود دارد، استفاده از تابع <code>()value_count</code> روی هر یک از ستون‌ها به طور مستقل است
<br>
در این حالت، تمامی مقادیر موجود در ستون رنگ‌ها را مشاهده می‌کنیم که تعداد هرکدام از آن‌ها شمره شده است. به این معنی که در دیتافریم ماشین‌ها، ما دو ماشین با رنگ قرمز و دو ماشین با رنگ سفید داریم.
</font>
</p>
