```python
from matplotlib import pyplot as plt

years = (2010, 2011, 2012, 2013, 2014, 2015, 2016)
salaries = (7800, 6700, 1120, 4400, 1000, 1999, 8000)
prices = (4500, 3300, 900, 3200, 990, 700, 7500)

plt.style.use('dark_background')
plt.plot(years, salaries, 'o:r', label='Salaries')
plt.plot(years, prices, color='b', linestyle='-.', marker='^', label='Prices')
plt.title('Salaries vs. Prices')
plt.xlabel('USD')
plt.ylabel('Year')
plt.legend()
plt.grid()
plt.savefig('/home/hossein/Personal/Learning/Learning_matplotlib/plots/one.png')
plt.show()
```
